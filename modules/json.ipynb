# json module

1. built-in package
2. work with JSON data

## importing

```python
import json
```

## json data

```python
x =  '''{
"name":"PONNAM SRI SAI", 
"age":18, 
"city":"Ongole",
"date":"24-1-2026",
"time":"00:14"
}'''
```

## json to python

```python
y = json.loads(x)
y, type(y)
```

## python to json

```python
y
```

```python
z = json.dumps(y)
print(z)
```

```python
# pretty print
z = json.dumps(y, indent=4)
print(z)
```

```python
# changing the separators
z = json.dumps(y, indent=4, separators=(". ", " = "))
print(z)
```

```python
# sorting keys
z = json.dumps(y, indent=4, sort_keys=True)
print(z)
```

```python
print(json.dumps({"name": "John", "age": 30}))
print(json.dumps(["apple", "bananas"]))
print(json.dumps(("apple", "bananas")))
print(json.dumps("hello"))
print(json.dumps(42))
print(json.dumps(31.768888888888888888888888888888))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

## customize encoding

```python
# json.dumps(1 + 2j) raises TypeError
def custom_json(obj):
    if isinstance(obj, complex):
        return {'__complex__': True, 'real': obj.real, 'imag': obj.imag}
    raise TypeError(f'Cannot serialize object of {type(obj)}')

json.dumps(1 + 2j, default=custom_json)
```

```python
class A:
    x = 22
    def __init__(self):
        self.y = 1

def serialize(obj):
    if isinstance(obj, A):
        return {
            "class": "A",
            "x": obj.x,
            "y": obj.y
        }
    raise TypeError(f"Object of type {type(obj).__name__} is not JSON serializable")

print(json.dumps(A(), default=serialize, indent=4))
print(json.dumps(A(), default=lambda o: o.__dict__, indent=4))
```

## customizing decoding

```python
def as_complex(dct):
    if '__complex__' in dct:
        return complex(dct['real'], dct['imag'])
    return dct

json.loads('{"__complex__": true, "real": 1, "imag": 2}',
    object_hook=as_complex)
```

```python
import decimal
json.loads('1.1', parse_float=decimal.Decimal)
```
