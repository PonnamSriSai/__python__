(Question 9 and 10)
Perform Hyperparameter Tuning using GridSearchCV on AdaBoostRegressor. Use random_state = 42. Value of cv = 4. Hyperparameter tuning is to be done over:

- n_estimators as [10,50,100,200,500]

- learning_rate as [0.1,0.5,1,2]

Use the best model obtained to compute the score on the test data



```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostRegressor
```

```python
?AdaBoostRegressor
```

```python
abr_model = AdaBoostRegressor(
    random_state = RANDOM_STATE
)
```

```python
?GridSearchCV
```

```python
best_abr = GridSearchCV(
    estimator = abr_model,
    param_grid = {
        'n_estimators': [10, 50, 100, 200, 500],
        'learning_rate': [0.1, 0.5, 1, 2]
    },
    cv = 4
)
```

```python
best_abr.fit(X_train, y_train)

y_pred = best_abr.predict(X_test)

r2_score(y_test, y_pred)
```

```python
# Define base model
ada_model = AdaBoostRegressor(random_state=42)

# Define parameter grid
param_grid = {
    'n_estimators': [10, 50, 100, 200, 500],
    'learning_rate': [0.1, 0.5, 1, 2]
}

# GridSearchCV
grid_search = GridSearchCV(
    estimator=ada_model,
    param_grid=param_grid,
    cv=4,
    scoring='r2'
)

# Fit on training data
grid_search.fit(X_train, y_train)

# Best model
best_model = grid_search.best_estimator_

# Predictions on test set
y_pred_ada = best_model.predict(X_test)

# Compute R² score
r2_ada = r2_score(y_test, y_pred_ada)

# Print results
print("Best Parameters:", grid_search.best_params_)
print("R2 Score (AdaBoost):", round(r2_ada, 3))
```

What is the value of n_estimators obtained for the best model after training with GridSearchCV?

```python
best_abr.best_params_["n_estimators"]
```

```python
# Extract best n_estimators from GridSearchCV
best_n_estimators = grid_search.best_params_['n_estimators']

print("Best n_estimators:", best_n_estimators)
```
