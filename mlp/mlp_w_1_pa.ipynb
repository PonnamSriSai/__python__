# 6

Apply the following functions to the columns Dep_Time and Arrival_Time:

Transform the values in the 'dep_time' and 'arrival_time' columns to represent the hour component. For instance, if an entry is 10:05 June 13 or 10:05, the corresponding value should be 10.

Then convert the time into four categories as follows:
$$
5 <= hour < 12 = Morning \\
12 <= hour < 17 = Afternoon \\
17 <= hour < 20 = Evening \\
20 <= hour < 5 = Night
$$

Note: Please ensure that you make the changes directly within the dataset and continue to use that modified dataset for subsequent questions.

How many flights started in the Morning and arrived the destination at Evening?

```python
df['hour_of_dep'] = df['dep_time'].str.extract(r'(\d+):').astype(int)
df['hour_of_arr'] = df['arrival_time'].str.extract(r'(\d+):').astype(int)
```

```python
df[['dep_time', 'hour_of_dep', 'arrival_time', 'hour_of_arr']].sample(5)
```

```python
bins = [0, 5, 12, 17, 20, 24]
labels = ['Night', 'Morning', 'Afternoon', 'Evening', 'Night']
df['cat_of_dep'] = pd.cut(
    df['hour_of_dep'],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True,
    ordered=False
)
df['cat_of_arr'] = pd.cut(
    df['hour_of_arr'],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True,
    ordered=False
)
df[['hour_of_dep', 'cat_of_dep', 'hour_of_arr', 'cat_of_arr']].sample(5)
```

```python
df[(df['cat_of_dep'] == 'Morning') & (df['cat_of_arr'] == 'Evening')].shape[0]
```

# 3

Is the average price of flight tickets higher on weekends (Saturday and Sunday) or on weekdays (Remaining 5 days)?



```python
df['weekend'] = df['weekday'].isin(['Saturday', 'Sunday'])
avg_price_weekend = df[df['weekend']]['price'].mean()
avg_price_weekday = df[~df['weekend']]['price'].mean()
avg_price_weekday, avg_price_weekend
```
