
Load the fetch_california_housing dataset from sklearn. Split the data using train_test_split with test_size = 0.2 and random_state as 1. Train the following models using the training data and calculate the R2 Score and Root Mean Squared Error (RMSE) on the test data.

- Model 1: Linear Regression

- Model 2: AdaBoost Regressor with random state = 1

- Model 3: Decision Tree Regressor with random state = 1


Choose the appropriate option:



(A) r2_score (Model 2) > r2_score (Model 1) > r2_score (Model 3)

(B) r2_score (Model 3) > r2_score (Model 1) > r2_score (Model 2)

(C) RMSE (Model 2) > RMSE (Model 1) > RMSE (Model 3)

(D) RMSE (Model 3) > RMSE (Model 1) > RMSE (Model 2)

```python
# Import required libraries
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Load dataset
data = fetch_california_housing()
X = data.data
y = data.target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)

# -----------------------------
# Model 1: Linear Regression
# -----------------------------
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

r2_lr = r2_score(y_test, y_pred_lr)
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))

# -----------------------------
# Model 2: AdaBoost Regressor
# -----------------------------
ada = AdaBoostRegressor(random_state=1)
ada.fit(X_train, y_train)
y_pred_ada = ada.predict(X_test)

r2_ada = r2_score(y_test, y_pred_ada)
rmse_ada = np.sqrt(mean_squared_error(y_test, y_pred_ada))

# -----------------------------
# Model 3: Decision Tree
# -----------------------------
dt = DecisionTreeRegressor(random_state=1)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

r2_dt = r2_score(y_test, y_pred_dt)
rmse_dt = np.sqrt(mean_squared_error(y_test, y_pred_dt))

# -----------------------------
# Print Results
# -----------------------------
print("R2 Scores:")
print(f"Linear Regression: {r2_lr:.4f}")
print(f"AdaBoost Regressor: {r2_ada:.4f}")
print(f"Decision Tree: {r2_dt:.4f}")

print("\nRMSE:")
print(f"Linear Regression: {rmse_lr:.4f}")
print(f"AdaBoost Regressor: {rmse_ada:.4f}")
print(f"Decision Tree: {rmse_dt:.4f}")

# -----------------------------
# Compare
# -----------------------------
print("\nComparisons:")
print("R2 ranking:", sorted(
    [("LR", r2_lr), ("ADA", r2_ada), ("DT", r2_dt)],
    key=lambda x: x[1], reverse=True
))

print("RMSE ranking:", sorted(
    [("LR", rmse_lr), ("ADA", rmse_ada), ("DT", rmse_dt)],
    key=lambda x: x[1]
))
```

Load the Iris Dataset from sklearn. Split the data using train_test_split with test_size = 0.33 and random_state as 1. Train the Logistic Regression model (with random state as 1) on the training data and make predictions on the test data. Print the confusion matrix and the classification report


How many samples has the model misclassified?

```python
# Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=1
)

# Train Logistic Regression model
model = LogisticRegression(random_state=1, max_iter=200)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Classification Report
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Misclassified samples
misclassified = np.sum(y_test != y_pred)
print("\nNumber of misclassified samples:", misclassified)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# 1. Load the dataset
iris = load_iris()
X, y = iris.data, iris.target

# 2. Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=1
)

# 3. Initialize and train the Logistic Regression model
# We increase max_iter to ensure convergence on this dataset
model = LogisticRegression(random_state=1, max_iter=200)
model.fit(X_train, y_train)

# 4. Make predictions
y_pred = model.predict(X_test)

# 5. Print results
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# 6. Calculate misclassified samples
misclassified = (y_test != y_pred).sum()
print(f"\nNumber of misclassified samples: {misclassified}")
```


Split the vectorized data and label into train and validation sets using train_test_split with test_size = 0.3 and random_state = 1. Train a MultinomialNB model on the training dataset and compute the score on the validation set. Enter the score obtained.


```python
# Import libraries
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

# Load TRAIN subset
X, y = fetch_20newsgroups(subset='train', return_X_y=True)

# Vectorize using TF-IDF
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X)

# Train-validation split
X_train, X_val, y_train, y_val = train_test_split(
    X_tfidf, y, test_size=0.3, random_state=1
)

# Train Multinomial Naive Bayes
model = MultinomialNB()
model.fit(X_train, y_train)

# Compute score (accuracy)
score = model.score(X_val, y_val)

print("Validation Accuracy:", round(score, 4))
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

# 1. Load the train subset
X_raw, y_raw = fetch_20newsgroups(subset='train', return_X_y=True)

# 2. Vectorize the data
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X_raw)

# 3. Split into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X_tfidf, y_raw, test_size=0.3, random_state=1
)

# 4. Train the Multinomial Naive Bayes model
clf = MultinomialNB()
clf.fit(X_train, y_train)

# 5. Compute and print the score
score = clf.score(X_val, y_val)
print(f"Validation Score: {score}")
```
